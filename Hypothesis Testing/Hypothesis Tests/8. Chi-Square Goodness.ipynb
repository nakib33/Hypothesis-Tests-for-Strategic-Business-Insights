# üöÄ**Project Title: Chi-Square Goodness of Fit Test on Customer Distribution Across Marketing Campaigns**

**Description:**

This project aims to determine whether the observed distribution of customers across three marketing campaigns (Email, Social Media, TV) significantly differs from the expected distribution. The Chi-Square Goodness of Fit test is a statistical method used here to test if the observed frequencies fit a hypothesized distribution.



## **üóÇÔ∏è 1: Import Necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chisquare
```

## **üéØ 2: Define Business Scenario & Goal**

**Scenario:**
A marketing team runs three types of campaigns (Email, Social Media, TV) and wants to verify if the distribution of customers responding to these campaigns differs from their expected proportions based on historical data.

**Goal:**
To test if the observed customer counts for each campaign are significantly different from the expected counts.

## **üîë 3: Define Hypotheses**

* **Null Hypothesis (H‚ÇÄ):** The observed customer distribution matches the expected distribution. (Observed = Expected)

* **Alternative Hypothesis (H‚ÇÅ):** The observed customer distribution does not match the expected distribution. (Observed ‚â† Expected)

## **üîó 4: Data Generation**

Generate synthetic data representing the observed counts of customers responding to each campaign. Also, define expected counts (from historical or theoretical knowledge).

```python
# Observed number of customers for each campaign (synthetic data)
observed_customers = np.array([120, 95, 85])  # Email, Social Media, TV

# Expected distribution based on historical data or business assumptions
expected_proportions = np.array([0.4, 0.35, 0.25])  # Expected % of customers per campaign

# Convert proportions to expected counts, total customers = sum of observed
total_customers = observed_customers.sum()
expected_customers = expected_proportions * total_customers

print(f"Observed counts: {observed_customers}")
print(f"Expected counts: {expected_customers.round(2)}")
```

## **‚ú® 5: Data Visualization**

```python
campaigns = ['Email', 'Social Media', 'TV']

plt.figure(figsize=(8,5))
bar_width = 0.35
indices = np.arange(len(campaigns))

plt.bar(indices, observed_customers, bar_width, label='Observed', color='skyblue')
plt.bar(indices + bar_width, expected_customers, bar_width, label='Expected', color='orange')

plt.xlabel('Campaign Type')
plt.ylabel('Number of Customers')
plt.title('Observed vs Expected Customer Counts Across Campaigns')
plt.xticks(indices + bar_width / 2, campaigns)
plt.legend()
plt.show()
```

## **üìí 6: Assumptions of Chi-Square Goodness of Fit Test**

* **Independence:** Customer counts across campaigns are independent observations.

* **Scale of Measurement**: Categorical data with counts/frequencies.

* **Expected Frequency:** Each expected frequency should generally be ‚â• 5 (satisfied here).

* **Normality & Homogeneity:** Not applicable as Chi-Square test is non-parametric and does not assume normality or homogeneity.



## **üë∑‚Äç‚ôÇÔ∏è 7: Perform Chi-Square Goodness of Fit Test**

```python
# Adjust expected counts to ensure sum equals observed sum due to rounding
expected_customers[-1] = total_customers - expected_customers[:-1].sum()

# Chi-Square Goodness of Fit Test
chi_stat, p_value = chisquare(f_obs=observed_customers, f_exp=expected_customers)

print(f"Chi-square Goodness of Fit Test: Statistic = {chi_stat:.4f}, p-value = {p_value:.4f}")
```

## **‚úîÔ∏è 8: Interpretation of Results**

```python
alpha = 0.05  # significance level

if p_value < alpha:
    print(f"Since p-value ({p_value:.4f}) < {alpha}, we reject the null hypothesis.")
    print("There is sufficient evidence to conclude that the observed customer distribution differs significantly from the expected distribution.")
else:
    print(f"Since p-value ({p_value:.4f}) ‚â• {alpha}, we fail to reject the null hypothesis.")
    print("There is insufficient evidence to conclude that the observed customer distribution differs from the expected distribution.")
```

**üîö Summary:**


The Chi-square Goodness of Fit test yielded a statistic of 2.2857 with a p-value of 0.3189. Since the p-value is greater than 0.05, we fail to reject the null hypothesis at the 5% significance level. This indicates that there is insufficient evidence to suggest a significant difference between the observed and expected customer distributions across the campaigns. Therefore, the observed customer responses align well with the expected distribution.
